```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

##IMPORT DATA read file
sale_data = pd.read_csv("/content/saledata.csv")
#print(sale_data)

##Data preprocessing
x=sale_data.iloc[:,:-1].values #[: all rows, : -1 all column except last]
y=sale_data.iloc[:,-1].values #[: all rows, -1 only last column
print(x)
print(y)

##splitting the datasets into the training set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

##implement model
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train, y_train)

#predict the model
y_pred=regression.predict(x_test)
tv_budget=float(input("Enter tv budget"))
np_budget=float(input("Enter news paper budget"))
rd_budget=float(input("Enter radio budget"))
testinput=[[tv_budget, np_budget, rd_budget]]
predicted_output=regression.predict(testinput)
print("sales unit=",predicted_output)

#graphical representation of data
for i in range(x_train.shape[1]):
    plt.scatter(x_train[:,i],y_train,color='red')

#plt.plot(x_train[:,1], regression.predict(x_train), color='blue')
plt.title("budegt vs Sales")
plt.xlabel("budget")
plt.ylabel("sales")
plt.show()

#model accuracy
from sklearn.metrics import mean_squared_error, mean_absolute_error
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)

#target
#rmse 70 to 150
#mae 50 to 100
#mse 5000 to 20000
```
