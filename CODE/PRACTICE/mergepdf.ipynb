```python
#code to decrypt pdf file and merge pdf files with known password
from pypdf import PdfWriter, PdfReader
pdf_directory = r"actual path of your directory"
pdf_files = sorted(
    os.path.join(pdf_directory, fn) 
    for fn in os.listdir(pdf_directory)
    if fn.lower().endswith(".pdf")
)
writer = PdfWriter()
for pdf in pdf_files:
    reader = PdfReader(pdf)
    if reader.is_encrypted:
        try:
            reader.decrypt("test")  # Attempt to decrypt with the correct password
        except Exception as e:
            print(f"Could not decrypt {pdf}: {e}")
            continue  # Skip this file if it cannot be decrypted
    for page in reader.pages:
        writer.add_page(page)   
with open("merged2.pdf", "wb") as f_out:
    writer.write(f_out)

```

```python
#code to decrypt pdf files with known password``
pdf_path = "actual path of your pdf file"  # Replace with your PDF file path
password = "test"  # Replace with the actual password

reader = PdfReader(pdf_path)
writer = PdfWriter()

if reader.is_encrypted:
    if reader.decrypt(password):
        for page in reader.pages:
            writer.add_page(page)
    else:
        print("Failed to decrypt PDF")
else:
    print("PDF is not encrypted")

with open("decrypted.pdf", "wb") as f_out:
    writer.write(f_out)
```
