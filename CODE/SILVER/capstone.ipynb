```python
import cv2
import torch
import matplotlib.pyplot as plt
from torchvision import transforms
```

```python
# Load your PyTorch fire detection model
loaded_data = torch.load("C:\\user1\\python\\GenAI\\CODE\\model\\Fire_Detection.pt", weights_only=False)

# Assuming the actual model is stored under the key 'model' in the loaded dictionary
if isinstance(loaded_data, dict) and 'model' in loaded_data:
    model = loaded_data['model']
elif isinstance(loaded_data, torch.nn.Module):
    model = loaded_data
else:
    raise TypeError("The loaded file does not contain a PyTorch model in an expected format.")

# Use CPU by default and convert the model to float32 for compatibility
device = torch.device("cpu")
model = model.to(device).float()
model.eval()
```

```python
# Define preprocessing steps
preprocess = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

```python
import os

# Specify the folder path containing your image frames
folder_path = 'C:\\user1\\python\\GenAI\\CODE\\images'

# List all files in the folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
image_files.sort()

frames = []
for image_file in image_files:
    img_path = os.path.join(folder_path, image_file)
    print(f"Processing image: {img_path}")
    img = cv2.imread(img_path)
    if img is not None:
        frames.append(img)

        # Convert BGR (OpenCV) to RGB for matplotlib display
        rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Preprocess image
        x = preprocess(rgb_img).unsqueeze(0).to(device)

        # Run inference
        model.eval()
        with torch.no_grad():
            output = model(x)

        # Some models return a tuple/list of tensors; use the first tensor-like output
        if isinstance(output, (tuple, list)):
            output = output[0]

        # Convert model output to a fire-confidence score
        if output.ndim == 0:
            confidence = float(torch.sigmoid(output))
        elif output.ndim == 2 and output.shape[1] >= 2:
            probs = torch.softmax(output, dim=1)
            confidence = float(probs[0, 1]) if probs.shape[1] > 1 else float(probs[0, 0])
        elif output.ndim == 1:
            confidence = float(torch.sigmoid(output[0])) if output.numel() == 1 else float(output[0])
        else:
            confidence = float(torch.sigmoid(output).mean())

        print(f"Image: {img_path}")
        print(f"Fire confidence: {confidence * 100:.2f}%")

        if confidence >= 0.5:
            print("Prediction: Fire detected")
            plt.figure(figsize=(5, 5))
            plt.imshow(rgb_img)
            plt.axis("off")
            plt.title(f"Fire confidence: {confidence * 100:.2f}%")
            plt.show()
            plt.close()
        else:
            print("Prediction: No fire detected")
            print("Image not shown because no fire was detected.")

print(f"Loaded {len(frames)} frames from {folder_path}")
```
