```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

##read data from CSV
salary_data = pd.read_csv("/content/salarydata_detailed.csv")
print(salary_data)
df = pd.DataFrame(salary_data, columns=['Age', 'Gender', 'Edu', 'Job', 'Exp', 'Salary'])

##encode string data to number
lab_encoder = LabelEncoder()
#df['Gender_ec', 'Edu_ec', 'Job_ec'] = lab_encoder.fit_transform(['Gender', 'Edu', 'Job']) #not working, need to encode each colum seperatly
df['Gender_ec'] = lab_encoder.fit_transform(df['Gender'])
df['Edu_ec'] = lab_encoder.fit_transform(df['Edu'])
df['Job_ec'] = lab_encoder.fit_transform(df['Job'])

#remove NnN values, or model will give error
df = df.dropna()
##extract independent and dependent variable
x = df[['Age', 'Gender_ec', 'Edu_ec', 'Job_ec', 'Exp']]
#x = df[['Age', 'Gender_ec', 'Exp']]
#x = df[['Exp']]
y = df[['Salary']]
print(x)
print(y)

import seaborn as sns
sns.pairplot(x)

##fit and transform
#scl = StandardScaler()
#x=scl.fit_transform(x)
#print(x)
```

```python
#split train and test dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
##implement linear model
regression = LinearRegression()
regression.fit(x_train, y_train)

## try ridge model
#from sklearn.linear_model import Ridge
# Ridge regression with regularization
#regression = Ridge(alpha=0.12)  # Adjust alpha (regularization strength)
#regression.fit(x_train, y_train)
```

```python
## Predict on the test data
y_pred = regression.predict(x_test)

# Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
err_mse = mean_squared_error(y_test, y_pred)
err_mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", err_mse)
print("Mean Absolute Error:", err_mae)
print("R-squared:", r2)

#target
#rmse 70 to 150
#mae 50 to 100
#mse 5000 to 20000
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) #fig size is size of plot in inches

# Plot data on the subplots
axes[0, 0].plot(x[:,0], y, 'r--') #'r.' print scatter and 'r--' plot line
axes[0, 0].set_title('Age vs Salary')
axes[0, 1].plot(x[:,1], y, 'g.')
axes[0, 1].set_title('Gender vs Salary')
axes[1, 0].plot(x[:,2], y, 'b.')
axes[1, 0].set_title('Exp vs Salary')
axes[1, 1].plot(x_test[:,2], y_pred, 'y.')
axes[1, 1].set_title('Output')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
